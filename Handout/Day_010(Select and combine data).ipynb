## Data filter

```python
# Search by condition 
iris[iris.sepal_length_in_cm < 4.5]
```

```python
# Search by condition 
iris.loc[iris.sepal_length_in_cm < 4.5]
```

## Combine Data

```python
df_1 = pd.DataFrame([[1, 2, 3], [4, 5, 6]], index=[0, 1], columns=['A', 'B', 'C'])
df_2 = pd.DataFrame([[7, 8, 9], [0, 1, 2]], index=[0, 1], columns=['B', 'C', 'D'])
pd.concat([df_1, df_2], axis=0)
```

```python
pd.concat([df_1, df_2], axis=1)
```

```python
pd.concat([df_1, df_2], axis=0, join='inner')
```

```python
pd.merge(df_1, df_2, on=['B', 'C'], how='outer')
```

```python
pd.merge(df_1, df_2, on=['B', 'C'], how='right')
```

```python
df_1_index = df_1.set_index(['B'])
df_2_index = df_2.set_index(['B'])
df_1_index.join(df_2_index, how='outer', lsuffix='_left', rsuffix='_right')
```
