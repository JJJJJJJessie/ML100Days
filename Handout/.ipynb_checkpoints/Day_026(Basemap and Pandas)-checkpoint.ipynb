## City population

```python
import numpy as np
import pandas as pd
import cartopy.crs as ccrs
from netCDF4 import Dataset
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset as NetCDFFile
```

```python
# Read file
posi=pd.read_csv("D:/Data Analysis/Python Data Science/ML100Days/Data/2014_us_cities.csv")
posi.head()
```

```python
posi.info()
```

```python
# Choose the first 280 cities' data
lat = np.array(posi["lat"][0:500]) # Get the lat vlaues
lon = np.array(posi["lon"][0:500]) # Get the lon values
pop = np.array(posi["pop"][0:500], dtype=float) # Get the population, turning to the float type

# Set the figure size
plt.figure(figsize=(8, 6))

# Found new map
map = Basemap(projection='stere',lat_0=90,lon_0=-105, llcrnrlat=23.41 ,urcrnrlat=45.44, 
              llcrnrlon=-118.67, urcrnrlon=-64.52, rsphere=6371200., resolution='l', area_thresh=10000)

# Draw
map.drawmapboundary()
map.drawstates() 
map.drawcoastlines() 
map.drawcountries() 
map.drawcounties() 

# Draw the scatter plot(Data type must be the float)
size=(pop/np.max(pop))*1000 

x,y = map(lon,lat)
map.scatter(x,y,s=size, color='orange')

plt.title('Population distribution in America')

# Show
plt.show()
```

```python
# Read file
data = Dataset('D:/Data Analysis/Python Data Science/ML100Days/Data/MERRA2_400.tavgM_2d_slv_Nx.202105.nc4', mode='r')

data.variables.keys()
```

```python
# Dtat settings
lon = data.variables['lon'][:]
lat = data.variables['lat'][:]
T2M = data.variables['T2M'][:,:,:]

# Get the time index
T2M = T2M[0,:,:]

# Settings of figure
fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection=ccrs.Robinson()) # Use the plt function on Robin projection
ax.set_global()
ax.coastlines(resolution="110m",linewidth=1)
ax.gridlines(linestyle='--',color='black')

# Draw the map
clevs = np.arange(230,311,5)
plt.contourf(lon, lat, T2M, clevs, transform=ccrs.PlateCarree(),cmap=plt.cm.jet)
plt.title('MERRA-2 202105', size=14)

# Set the color bar
# plt.color(pad-> interval, aspect-> width of color bar, shrink-> length of color bar)
cb = plt.colorbar(ax=ax, orientation="vertical", pad=0.04, aspect=20, shrink=0.7)
cb.set_label('K',size=12, rotation=0, labelpad=15)
cb.ax.tick_params(labelsize=10) # Color bar tick labels

# Save the image
fig.savefig('D:/Data Analysis/Python Data Science/ML100Days/Data/MERRA2.png', format='png', dpi=1200)
```
