```python
# Found new map
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)

# Color the whole Earth
map.drawmapboundary(fill_color = 'lightblue')

# Color the continent
map.fillcontinents(color = 'orange', lake_color = 'lightblue')

# Color the coastlines
map.drawcoastlines(color = 'black')

# Show
plt.show()
```

```python
# Found new map
map = Basemap(projection='merc',urcrnrlat=80, llcrnrlat=-80,llcrnrlon=-180, urcrnrlon=180)

# Draw
map.drawcoastlines()
map.drawparallels(range(-90,91,30), color='#CCCCCC')
map.drawmeridians(range(-180,181,60), color='#CCCCCC')

# Color the states
map.drawstates(color='white') 

# Color the countries
map.drawcountries(color='white')

# Color the orean
map.drawmapboundary(fill_color='lightblue')

# Color the continents
map.fillcontinents(color = 'orange')

# Show
plt.show()
```

```python
import numpy as np
from itertools import chain

# Found the new map
fig = plt.figure(figsize=(8, 6), edgecolor='w')
m = Basemap(projection='cyl', resolution=None, llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180)

# Draw the shaded relief
m.shadedrelief(scale=0.2)

# Return the lats and lons in dict way
lats = m.drawparallels(np.linspace(-90, 90, 13))
lons = m.drawmeridians(np.linspace(-180, 180, 13))

# Use plt.Line2D to set the lats and lons
lat_lines = chain(*(tup[1][0] for tup in lats.items()))
lon_lines = chain(*(tup[1][0] for tup in lons.items()))
all_lines = chain(lat_lines, lon_lines)

# Setting the line
for line in all_lines:
    line.set(linestyle='-', alpha=0.3, color='w')

# Show
plt.show()
```

```python
def draw_map(m, scale=0.2):
    
    # Draw the shaded relief
    m.shadedrelief(scale=scale)

    # Return the lats and lons in dict way
    lats = m.drawparallels(np.linspace(-90, 90, 13))
    lons = m.drawmeridians(np.linspace(-180, 180, 13))

    # Use plt.Line2D to set the lats and lons
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)

    # Setting the line
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')
    
    # Show
    plt.show()
```

```python
# Found the new map
fig = plt.figure(figsize=(8, 8), edgecolor='w')
m = Basemap(projection='moll', resolution=None, lat_0=0, lon_0=0)

# Show
draw_map(m)
```
