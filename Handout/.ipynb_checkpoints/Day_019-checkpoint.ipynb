## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y = np.sin(x)

plt.plot(x, y)
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)

plt.plot(x, y, 'go-', label='line 1', linewidth=2)
```

```python
x = np.arange(10)

plt.plot(x,x*1.5,'go:',x,x*2.5,'rx',x,x*3.5,'^',x,x*4.5,'bd-.')
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)

plt.figure(figsize=(8,6), facecolor='pink')

plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plot with figsize (8,6)")
plt.show()
```

## Subplot

```python
# Found 2X2 windows
fig, axes = plt.subplots(2, 2)

# Data
data = pd.Series(np.random.rand(5), index=list('12345'))

# ax is location, alpha is transparency 
data.plot.bar(ax=axes[0,1], color='y', alpha = 0.5) 
data.plot.barh(ax=axes[1,1], color='g', alpha=0.3)
data.plot.bar(ax=axes[1,0], color='b', alpha = 0.3) 
data.plot.barh(ax=axes[0,0], color='r', alpha=0.3)
 
plt.show()
```

## Scatter Plots

```python
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.scatter(X, Y, color='y', alpha=0.5, s=100, edgecolors='k')
plt.title("Scatter plot ")
```
