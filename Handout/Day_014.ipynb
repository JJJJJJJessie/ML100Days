```python
import sqlite3 

# Read excel file
iris_data = pd.read_excel('iris.xlsx', sheet_name='iris')
connection = sqlite3.connect('sql_db.sqlite')

# Found SQL file
iris_data.to_sql('iris_SQL', connection, if_exists='replace')
connection.commit()
connection.close()

# Read SQL file
connection = sqlite3.connect('sql_db.sqlite')
iris_data_sql = pd.io.sql.read_sql("SELECT * FROM iris_SQL", connection)
connection.close()
iris_data_sql
```

## Fill missing value (NaN)

```python
temp_data = pd.DataFrame([['2020-11-01', 24.8],
             ['2020-11-02', 24.8],
             ['2020-11-03', None],
             ['2020-11-04', 25]],columns=['date','current_temp'])
temp_data
```

```python
# Fillna()
temp_data.fillna(0)
```

```python
temp_data.fillna(temp_data['current_temp'].median())
```

```python
# Fill data by previous one
temp_data.fillna(method='ffill')
```

```python
# Fill data by later one
temp_data.fillna(method='bfill')
```
