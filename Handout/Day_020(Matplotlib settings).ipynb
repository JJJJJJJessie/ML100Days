## Found Subplots / Axes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the figure size
fig = plt.figure(figsize=(10,6))

# Found three subplots
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

# Show
plt.xlabel("x-axis") 
plt.show()
```

```python
# Set the values
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# First window
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("Sin()")

# Second window
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("Cos()")

# Show
plt.show()
```

```python
from sklearn import datasets

# Load 'digits'
digits = datasets.load_digits()

# Set figure size
fig = plt.figure(figsize=(6, 4))

# Set subplots
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# Show the number in each window
for i in range(8):
    ax = fig.add_subplot(2, 4, i + 1, xticks = [], yticks = []) # Close the ticks
    ax.imshow(digits.images[i], cmap = plt.cm.binary)
    ax.text(0, 7, str(digits.target[i])) # Num

# Show
plt.show()
```

```python
# Outer window
plt.axes([0.1,0.1,.8,.8])
plt.xticks([]), plt.yticks([])
plt.text(0.6, 0.6, 'axes([0.1,0.1,.8,.8])', ha='center', va='center', size=20, alpha=.6)

# Inner window
plt.axes([0.2,0.2,.3,.3])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'axes([0.2,0.2,.3,.3])', ha='center', va='center', size=12, alpha=.5)

# Show
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Found coordinate
fig = plt.figure()
ax = Axes3D(fig)

# Search settings
# help(plt.plot)
# help(np.random.sample)

# Draw sin()
x = np.linspace(0, 1, 100)
y = np.cos(x * 2 * np.pi) / 2 + 0.5

# Draw the line on z-axis(zdir = 'z') which z = 0.5.
ax.plot(x, y, zs = 0.5, zdir = 'z', color = 'black', label = 'curve in (x, y)')

# Show
plt.show()
```

```python
# Draw multiple lines
fig = plt.figure(figsize = (12,8)) ; ax = fig.add_subplot(1,1,1)

ax.plot ( np.random.randn(1000).cumsum(),'k' , label = 'one')
ax.plot ( np.random.randn(1000).cumsum(),'k--' , label = 'two')
ax.plot ( np.random.randn(1000).cumsum(),'k.' , label = 'three')

ax.legend(loc = 'best')
plt.show()
```
