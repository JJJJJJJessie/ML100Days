## Negative Binomial Distribution

```python
import math
import statistics
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Define the data
p = 0.4
k = 3 # When success 3 times
x = np.arange(stats.nbinom.ppf(0.01, k, p), stats.nbinom.ppf(0.99, k, p)) 
print(x)

# Calculate p(X=x)
# Calculate the probability mass function(pmf)
# f(x,k,p) = [choose(x+k-1, k-1) * p**(k-1) * (1-p)**(x)] * p ; n = x+k
probs = stats.nbinom.pmf(x, k, p) # 'x' is The number of failure, 'k' is the nummber of success.

# Plot
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('NB(k=3,p=0.4)')
plt.show()
```

```python
# Calculate p(X<=x)
# Calculate the cumulative density function(cdf)
cumsum_probs = stats.nbinom.cdf(x, k, p)

# Plot
plt.show()
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('binomial(n=5,p=0.4)')
plt.plot(x, cumsum_probs)
plt.show()
```

```python
# Calculate the percent point function(ppf, which is the inverse of the cdf)
p_loc= stats.nbinom.ppf(cumsum_probs, k, p)
print(p_loc)
```

```python
# Run the random sample
X = stats.nbinom.rvs(k, p, size=200)

plt.hist(X)
plt.show()
```

```python
# Calculate the mean, variance, kurtosis and skew(mvks)
stat_nbin=stats.nbinom.stats(k, p, moments='mvks')
print(stat_nbin)
print(type(stat_nbin))
print("negative binomial mean=",float(stat_nbin[0]))
print("negative binomial variance=",float(stat_nbin[1]))
print("negative binomial kurtosis=",float(stat_nbin[2]))
print("negative binomial skew=",float(stat_nbin[3]))
```

## Hypergeometric Distribution

```python
# Define the data
N = 50 # Total number
n = 10 # Extract number
K = 30 # Total number of objective things
x = np.arange(0, min(n+1,K+1))
print(x)

# Calculate p(X=x)
# Calculate the probability mass function(pmf)
# f(x,k,p) = [choose(K, x) * choose(N-K, n-x)] / choose(N, n)
probs = stats.hypergeom.pmf(x, N, K,n)

# Plot
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('pmf of Hypergeometric(N=50,K=30,n=10)')
plt.show()
```

```python
# Calculate p(X<=x)
# Calculate the cumulative density function(cdf)
cumsum_probs = stats.hypergeom.cdf(x, N,K,n)

# Plot
plt.show()
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of Hypergeometric(N=50,K=30,n=10)')
plt.plot(x, cumsum_probs)
plt.show()
```

```python
# Calculate the percent point function(ppf, which is the inverse of the cdf)
p_loc= stats.hypergeom.ppf(cumsum_probs, N, K, n)
print(p_loc)
```

```python
# Run the random sample
X = stats.hypergeom.rvs(N, K, n, size=2000)

plt.hist(X)
plt.show()
```

```python
# Calculate the mean, variance, kurtosis and skew(mvks)
stat_hyperg=stats.hypergeom.stats(N,K,n,moments='mvks')
print(stat_hyperg)
print(type(stat_hyperg))
print("negative hypergeom mean=",float(stat_hyperg[0]))
print("negative hypergeom variance=",float(stat_hyperg[1]))
print("negative hypergeom kurtosis=",float(stat_hyperg[2]))
print("negative hypergeom skew=",float(stat_hyperg[3]))
```
