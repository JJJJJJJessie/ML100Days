## Iris data

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load data
df = sns.load_dataset('iris')
df
```

```python
df.info()
```

```python
# Turn on horizontal
sns.boxplot(data = df, orient = "h")
```

```python
# stripplot
sns.stripplot(x = "species", y = "petal_length", data = df, jitter = .1)
```

```python
# Use with boxplot and stripplot
sns.boxplot(x = "species", y = "petal_length", data = df)
sns.stripplot(x = "species", y = "petal_length", data = df, jitter = .1)
```

## Kernel Density Estimates

```python
sns.set_style("ticks")

# "hue" is classification, "diag_kind" is about the diag, "kind" is about the not_diag.
sns.pairplot(df, hue='species', diag_kind="kde", kind="scatter", palette="husl")
```

```python
sns.set_style("ticks")

# "hue" is classification, "diag_kind" is about the diag, "kind" is about the not_diag.
sns.pairplot(df, hue='species', diag_kind="hist", kind="reg", palette="husl")
```

```python
# Customize
g = sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
