## Discrete Uniform Distribution

```python
import math
import statistics
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Define the data
low = 1 
high = 7
x = np.arange(low,high)

# Calculate p(X=x)
# Calculate the probability mass function(pmf)
probs = stats.randint.pmf(x, low, high)
print('Probs:', probs)

# Plot
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('pmf of DU(1,6)')
plt.show()
```

```python
# Calculate p(X<=x)
# Calculate the cumulative density function(cdf)
cumsum_probs = stats.randint.cdf(x, low,high)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of DU(1,6)')
plt.plot(x, cumsum_probs)
plt.show()
```

```python
# Calculate the percent point function(ppf, which is the inverse of the cdf)
k = stats.randint.ppf(cumsum_probs, low, high)
print(k)
```

```python
# Run the random sample
plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
X = stats.randint.rvs(low,high,size=20)
plt.hist(X, bins=15)
plt.title('sample size = 20')
plt.show()

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 2)
X = stats.randint.rvs(low,high,size=2000)
plt.hist(X, bins=15)
plt.title('sample size = 2000')
plt.show()
```

```python
# Calculate the mean, variance, kurtosis and skew(mvks)
stat_randint = stats.randint.stats(low, high, moments='mvks')
print(stat_randint)
print(type(stat_randint))
print("randint mean=", float(stat_randint[0]))
print("randint variance=", float(stat_randint[1]))
print("randint kurtosis=", float(stat_randint[2]))
print("randint skew=", float(stat_randint[3]))
```

## Bernoulli Distribution

```python
# Define the prob of event
p = 0.4
x = np.arange(0, 2)

# Calculate p(X=x)
# Calculate the probability mass function(pmf)
probs = stats.bernoulli.pmf(x, p)
print('Probs:', probs)

# Plot
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('Bernoulli(p=0.4)')
plt.show()
```

```python
# Calculate p(X<=x)
# Calculate the cumulative density function(cdf)
cumsum_probs = stats.bernoulli.cdf(x,p)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('bernoulli(p=0.4)')
plt.plot(x, cumsum_probs)
plt.show()
```

```python
# Calculate the percent point function(ppf, which is the inverse of the cdf)
p_loc = stats.bernoulli.ppf(cumsum_probs, p)
print(p_loc)
```

```python
# Run the random sample
X = stats.bernoulli.rvs(p,size=20)
print(X)
plt.hist(X)
plt.show()
```

```python
# Calculate the mean, variance, kurtosis and skew(mvks)
stat_ber=stats.bernoulli.stats(p,moments='mvks')
print(stat_ber)
print(type(stat_ber))
print("bernoulli mean=",float(stat_ber[0]))
print("bernoulli variance=",float(stat_ber[1]))
print("bernoulli kurtosis=",float(stat_ber[2]))
print("bernoulli skew=",float(stat_ber[3]))
```

## Binomial Distribution

```python
# f(n,x,p) = choose(n, x) * p**x * (1-p)**(n-x)
# Define the data
p = 0.5
n = 5
x = np.arange(0, 6)

# Calculate p(X=x)
# Calculate the probability mass function(pmf)
probs = stats.binom.pmf(x, n, p)

# Plot
plt.bar(x, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n=5,p=0.5)')
plt.show()
```

```python
# Calculate p(X<=x)
# Calculate the cumulative density function(cdf)
cumsum_probs = stats.binom.cdf(x, n, p)

plt.show()
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('binomial(n=5,p=0.4)')
plt.plot(x, cumsum_probs)
plt.show()
```

```python
# Run the random sample
X = stats.binom.rvs(n, p, size=200)

plt.hist(X)
plt.show()
```

```python
# Calculate the mean, variance, kurtosis and skew(mvks)
stat_bin=stats.binom.stats(n,p,moments='mvks')
print(stat_bin)
print(type(stat_bin))
print("binomial mean=",float(stat_bin[0]))
print("binomial variance=",float(stat_bin[1]))
print("binomial kurtosis=",float(stat_bin[2]))
print("binomial skew=",float(stat_bin[3]))
```
