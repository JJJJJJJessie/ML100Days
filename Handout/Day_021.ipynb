## Seaborn Image

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset
data = np.array([3, 12, 5, 18, 45])
bars = np.array(['A', 'B', 'C', 'D', 'E'])
y_pos = np.arange(len(bars))

# Set y-axis
plt.yticks([3,5,12,18,45])

# Color is RGB format, and the last one is Alpha.
sns.barplot(x=y_pos, y=data, color=(0.2, 0.8, 0.6, 0.6)) 

# Add legend on left-upp
plt.legend(['b'])

# Customize axis title
plt.xlabel('title of the xlabel', fontweight='bold', color = 'red', alpha=0.5, fontsize='16', horizontalalignment='center')
plt.ylabel('title of the ylabel ',fontweight='bold', color = 'blue', alpha=0.5, fontsize='10', horizontalalignment='center')

# Background(darkgrid/whitegrid/dark/white/ticks)
sns.set_style('ticks')

# Adjust the figure size
plt.figure(figsize=(15, 10))

# Show
plt.show()
```

## Linear relationship

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

# Background(darkgrid/whitegrid/dark/white/ticks)
sns.set_style('whitegrid')

# Show
sinplot()
```

## Univariate analysis - Kernel Density Estimation(KDE)

```python
from scipy import stats, integrate 

x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)
kernels = []

for x_i in x:
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color="r")
    
sns.rugplot(x, color=".3", linewidth=3)
```

```python
from scipy.integrate import trapz

density = np.sum(kernels, axis=0)
density /= trapz(density, support)
plt.plot(support, density)
```

## Bivariate analysis
