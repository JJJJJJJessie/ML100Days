## Duplicated row

```python
# Check the number of duplicated rows
df_red.duplicated().sum(),df_white.duplicated().sum()
```

```python
# Drop the duplicated rows
df_red.drop_duplicates(inplace=True)
df_white.drop_duplicates(inplace=True)
```

```python
# Check if there has any null rows
df_red.isnull().sum().sum(), df_white.isnull().sum().sum()
```

```python
df_wine = df_red.append(df_white)
df_wine.to_csv('D:/Data Analysis/Python Data Science/ML100Days/Data/winequality.csv', index=False)
df_wine
```

## Histgram

```python
df_wine.hist(bins=10, color='lightblue',edgecolor='white',xlabelsize=8, ylabelsize=8, grid=False)

# plt.tight_layout(w_pad, h_pad, rect)
plt.tight_layout(rect=(1,1,3,3))
```

## Heatmap

```python
f, ax= plt.subplots(figsize=(10, 6))

# sns.heatmap(data, annot->write the value in the image or not, linewidths->the width of the grids, ax->coordinate)
b = sns.heatmap(df_wine.corr(), annot=True, linewidths=.05, ax=ax)

# Set the image_top(how long the image)
f.subplots_adjust(top=1)

# Get the y-axis value
bottom, top = ax.get_ylim()

# Set the new boundary
ax.set_ylim(bottom + 0.5, top - 1.5)

# Set the title
title= f.suptitle('Correlation Heatmap for wine attributes', fontsize=12)
```
