## DataFrrame calculation

```python
import pandas as pd

df_1 = pd.DataFrame([[1, 2, 3]], columns=['A', 'B', 'C'])
df_2 = pd.DataFrame([[4, 5, 6]], columns=['B', 'C', 'D'])

print(df_1 + 1)
print("")
print(df_1 + df_2)
```

## Sort DataFrame

```python
df_3 = pd.DataFrame({
    'col1': [0, 4, 2, 7, 3, 7],
    'col2': [2, 1, 2, 3, 1, 8]},
    index = ['A', 'a', 'B', 'B', 'B', 'b'])
df_3
```

```python
df_3.sort_values(by=['col1'])
```

```python
df_3.sort_values(by=['col1', 'col2'])
```

```python
df_3.sort_values(by=['col2'], ascending=False)
```

## Statistics

```python
df_3['col3'] = [4, 1, 5, 2, 3, 9]
df_3
```

```python
# Max
print(df_3.max())
print(df_3.col2.idxmax()) # Return the index of max_vlaue
```

```python
# Min
print(df_3.min())
print(df_3.col3.idxmin()) # Return the index of min_vlaue
```

```python
# Count numbers
df_3.count(axis=0)
```

```python
df_3.apply(lambda x : x**(0.5)*10)
```

```python
df_3['mean'] = round(df_3.mean(axis=1))
df_3
```

```python
df_3.loc['Mean'] = round(df_3.mean(axis=0))
df_3
```

```python
df_3.loc['std'] = round(df_3.std(axis=0))
df_3.loc['var'] = round(df_3.var(axis=0))
df_3
```

```python
df_3['sum'] = df_3.sum(axis=1)
df_3
```

```python
print(df_3['sum'].median())
print("")
print(df_3.median())
print("")
print(df_3.quantile(0.5))
```

```python
df_3.corr()
```
