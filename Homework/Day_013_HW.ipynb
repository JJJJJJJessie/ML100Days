## 1. 請接下列資料依照指定規則做合併：

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})

df1
```

```python
df2
```

```python
# According to 'fruit'
pd.merge(df1, df2, on='fruit')
```

```python
# According to index
df1.join(df2)
```

## 2. 承上題，請問為什麼依照 merge 合併後有些資料不見了？

```python
# According to 'fruit' by outer
pd.merge(df1, df2, on='fruit', how='outer')
```

## 3. 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

Becaues of the overlap columns.

```python
# According to index
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```
