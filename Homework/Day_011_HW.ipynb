## 1. 根據題目給的 DataFrame 完成下列操作：

```python
import numpy as np
import pandas as pd

data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)
df
```

```python
# Calcul the average age by all animal
df.age.mean()
```

```python
# Calcul the average age by same animal
df['mean_age'] = df.groupby('animal')['age'].transform('mean')
df
```

```python
# Sort by age & visits
df.sort_values(by=['age', 'visits'], ascending=[True, False])
```

```python
# Replace values
df['priority'] = df['priority'].replace(['yes', 'no'], [True, False])
df
```

## 2. 一個包含兩個欄位的 DataFrame，將每個數字減去：該欄位的平均數 / 該筆資料平均數

```python
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
```

```python
# Solution 1
df.sub(df.mean(), axis=1)
```

```python
# Solution 2
df.loc['mean_row'] = df.mean()
df.sub(df.loc['mean_row'], axis=1)
```

```python
# Solution 1
df.sub(df.mean(axis=1), axis=0)
```

```python
df['mean_col'] = df.mean(axis=1)
df.sub(df['mean_col'], axis=0)
```

## 3. 承上題，請問：哪一比的資料總合最小 / 哪一欄位的資料總合最小

```python
df['sum_col'] = df.sum(axis=1) - df['mean_col']
df.loc['sum_row'] = df.sum(axis=0) - df.loc['mean_row']
df
```

```python
df['sum_col'].idxmin()
```

```python
df.loc['sum_row'].idxmin()
```
