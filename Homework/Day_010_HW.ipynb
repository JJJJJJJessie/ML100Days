## 3. 請根據提供的資料，選擇出下列的要求：

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

```python
# filtered by first column > 20
df[df[0]>20]
```

```python
# filtered by first column + second column > 50

# Solution 1
df['0+1'] = df[0] + df[1]
df[df['0+1']>50]
```

```python
# Solution 2
df[(df.iloc[:, 0] + df.iloc[:, 1])>50]
```

```python
# filtered by first column < 30 or second column > 30

# Solution 1
df[(df[0]<30) & (df[1]>30)]
```

```python
# Solution 2
df[(df.iloc[:, 0]<30) & (df.iloc[:, 1]>30)]
```

```python
# filtered by total sum of row > 100
df['sum'] = df[0] + df[1] + df[2] + df[3]
df[df['sum']>100]
```

```python
df[df.sum(axis = 1)-df['sum']-df['0+1'] > 100]
```
