## 1. "Projection"參數為"ortho"，並以NASA寶石藍輸出圖像。

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# Set figure size
plt.figure(figsize=(6, 6))

# Found new map
m = Basemap(projection='ortho',lat_0=24, lon_0=121, resolution=None)

# Draw
m.bluemarble(scale=0.5)

# Show
plt.show()
```

## 2. "Projection"參數為"Mill"，所得平展位面。

```python
# Set figure size
plt.figure(figsize=(8, 6))

# Found new map
m = Basemap(projection='mill',llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='i')

# Draw
m.drawcoastlines()
m.drawstates(color='b')
plt.title('Basemap Tutorial')
m.drawcountries(linewidth=2, color='darkred')

# Show
plt.show()
```

## 3. "Projection"參數為"Icc"，並以NASA寶石藍輸出。

```python
plt.figure(figsize=(8, 6))

# Found new map
m = Basemap(projection='lcc', width=12000000,height=9000000, resolution=None, lat_0=20, lat_1=30,lon_0=110, lon_1=130)

# Draw
m.bluemarble(scale=0.5)

# Show
plt.show()
```
