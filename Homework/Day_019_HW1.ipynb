## 1. 畫出COS圖。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

plt.plot(x,y)
plt.show() 

# Sin()
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)
plt.show()

# Cos()
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.show()
```

## 2. 畫出Scatter Plots，顏色由（X，Y）的角度給出。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

colors = np.arctan2(Y,X)
plt.scatter(X, Y, s=25, c=colors, alpha=0.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
