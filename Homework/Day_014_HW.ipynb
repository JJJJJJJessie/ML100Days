## 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並依熱門程度排序後存成 csv 檔。

```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)

# Transfrom to dataframe
dcard = pd.DataFrame(data)
dcard.head(3)
```

```python
dcard = dcard.sort_values('subscriptionCount', ascending=False)
dcard.head(3)
```

```python
dcard.to_csv('Dcard.csv')
dcard.head(3)
```
