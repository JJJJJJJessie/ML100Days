## 1. 觀察 Age 和 Pclass 和 Sex 是否有關連性？

```python
import math
import statistics
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df_train = pd.read_csv("D:/Data Analysis/Python Data Science/ML100Days/Data/Titanic_train.csv")
df_train.head(5)
```

```python
# Check the missing values' rate
missing_vals = df_train.isnull().sum()/len(df_train)
missing_vals.sort_values(ascending=False)
missing_vals = pd.DataFrame(missing_vals, columns=['missing_rate'])
display(missing_vals)
```

```python
g = sns.FacetGrid(df_train, col='Pclass', hue='Sex')
g.map(sns.histplot, 'Age', edgecolor='white')
plt.legend()
```

## 2. 運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
from sklearn import preprocessing

# Transform sex to 0 and 1
le = preprocessing.LabelEncoder()
df_train['Sex'] = le.fit_transform(df_train['Sex'])
display(df_train)
```

```python
df_new = df_train[['Age', 'Pclass', 'Sex']]
df_new
```

```python
from sklearn.metrics.pairwise import nan_euclidean_distances

# Calculate the distances between each dot
display(pd.DataFrame(nan_euclidean_distances(df_new)))
```

```python
from sklearn.impute import KNNImputer

# Fill the values
value_neighbors = 3
imputer = KNNImputer(n_neighbors=value_neighbors, weights='distance')
df_filled = pd.DataFrame(imputer.fit_transform(df_new))
df_filled.rename(columns={0:'Age',1:'Pclass',2:'Sex'}, inplace=True)
display(df_filled)
```

```python
# Check the missing values' rate
missing_vals = df_filled.isnull().sum()/len(df_filled)
missing_vals.sort_values(ascending=False)
missing_vals = pd.DataFrame(missing_vals, columns=['missing_rate'])
display(missing_vals)
```
